Label Encoding
- converting the labels into numeric form

```python
# importing the dependencies

import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label encoding of Breast cancer dataset

```python
# loading the data from csv file to pandas dataframe
canser_data = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
# first 5 rows of the dataframe

canser_data.head()
```

```python
# finding the count of different labels

canser_data['diagnosis'].value_counts()
```

```python
# load label encoder function

label_ecode = LabelEncoder()
```

```python
labels = label_ecode.fit_transform(canser_data.diagnosis)
```

```python
# appending the labels to the dataframe

canser_data['target'] = labels
```

```python
canser_data.head()
```

- 0 ---> Benign
- 1 ---> Malignant

```python
canser_data['target'].value_counts()
```

Label encoding of iris data
